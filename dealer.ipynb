# Dealer Performance Analysis

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt 

# Loading and transforming data
df = pd.read_csv('/Users/abay/Downloads/Car Sales.xlsx - car_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.head()
df.Dealer_Region.value_counts()
```

# Total sales

```python
df_sum = df.groupby('Dealer_Region')['Price ($)'].sum().sort_values(ascending=False).reset_index()
df_sum
```

***The data shows that Austin leads in total sales with over $117 million, followed by Janesville and Scottsdale. These regions are likely major markets with strong demand. Greenville, Pasco, and Middletown also perform well but with slightly lower sales.***

# Average Price per Sale:
Determine the average price of cars sold by each dealer and region.
This helps to identify whether a dealer is focusing on high-end or budget-friendly vehicles.

```python
# per region 
df_avg_sale_region = df.groupby('Dealer_Region')['Price ($)'].mean().sort_values(ascending=False).reset_index()
df_avg_sale_region
```

```python
# per dealer
df_avg_sale_dealer = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending=False).reset_index()
df_avg_sale_dealer
```

# Top Selling Models:
Identify which car models are the most popular for customers.

```python
df_top_models = df.groupby('Model')['Car_id'].agg('size').sort_values(ascending=False)
df_top_models = df_top_models.head(10)

# plotting
fig, ax = plt.subplots(1,1)

df_top_models.plot(kind='barh')
ax.invert_yaxis()
ax.set_xlim(0,500)
ax.set_ylabel('Model')
ax.set_xlabel('Purchased')

fig.suptitle('Top Selling Models')
plt.tight_layout()
```
